```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('raman.csv')
path = '/Users/hailey/OneDrive/postech/0 개인미팅/03 정우철교수님'
```

```python
figx = 4.0
figy = 2.8
xmin = 150
xmax = 750
```

```python
df_pristine = df[df['model'] == 'pristine']
# print(df_pristine)
ymax = 3e15
plt.figure(figsize=(figx, figy))
# plt.axhline(y=0, color='k')
for i, row in df_pristine.iterrows():
    if int(row["freq-dft"]) == 362:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='k')
        plt.scatter(row['freq-dft'], row['activity'], color='k', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='left', va='bottom', rotation=45)
    elif row['activity'] > ymax/6:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='k')
        plt.scatter(row['freq-dft'], row['activity'], color='k', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='right', va='top', rotation=45)
    else:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='silver', zorder=0)
        plt.scatter(row['freq-dft'], row['activity'], color='silver', marker='D', s=10, zorder=0)
plt.xlabel('Frequency (cm$^{-1}$)', fontsize=11)
plt.ylabel('Activity (a.u.)', fontsize=11)
plt.xlim(xmin, xmax)
plt.ylim(0, ymax)
plt.xticks(range(xmin, xmax+1, 100))
plt.yticks(range(0, int(ymax)+1, int(ymax/6)))
plt.tight_layout()
plt.savefig(f'{path}/raman_plot_pristine.png', transparent=True)
plt.show()
```

```python
df_vacancy = df[df['model'] == 'vacancy']
# print(df_vacancy)
ymax = 3e14
plt.figure(figsize=(figx, figy))
# plt.axhline(y=0, color='k')
for i, row in df_vacancy.iterrows():
    if int(row["freq-dft"]) == 677:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='tab:orange')
        plt.scatter(row['freq-dft'], row['activity'], color='tab:orange', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='right', va='top', rotation=45)
    elif row['activity'] > ymax/6:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='tab:orange')
        plt.scatter(row['freq-dft'], row['activity'], color='tab:orange', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='left', va='bottom', rotation=45)
    else:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='silver', zorder=0)
        plt.scatter(row['freq-dft'], row['activity'], color='silver', marker='D', s=10, zorder=0)
plt.xlabel('Frequency (cm$^{-1}$)', fontsize=11)
plt.ylabel('Activity (a.u.)', fontsize=11)
plt.xlim(xmin, xmax)
plt.ylim(0, ymax)
plt.xticks(range(xmin, xmax+1, 100))
plt.yticks(range(0, int(ymax)+1, int(ymax/6)))
plt.tight_layout()
plt.savefig(f'{path}/raman_plot_vacancy.png', transparent=True)
plt.show()
```

```python
df_proton = df[df['model'] == 'proton']
# print(df_proton)
ymax = 6e11
plt.figure(figsize=(figx, figy))
# plt.axhline(y=0, color='k')
for i, row in df_proton.iterrows():
    if int(row["freq-dft"]) == 272 and row['activity'] > ymax/7:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='tab:blue')
        plt.scatter(row['freq-dft'], row['activity'], color='tab:blue', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='right', va='top', rotation=45)
    elif row['activity'] > ymax/6:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='tab:blue')
        plt.scatter(row['freq-dft'], row['activity'], color='tab:blue', marker='D', s=10)
        plt.annotate(f'{row["freq-dft"]:.0f}', (row['freq-dft'], row['activity']), 
        ha='left', va='bottom', rotation=45)
    else:
        plt.plot([row['freq-dft'], row['freq-dft']], [0, row['activity']], color='silver', zorder=0)
        plt.scatter(row['freq-dft'], row['activity'], color='silver', marker='D', s=10, zorder=0)
plt.xlabel('Frequency (cm$^{-1}$)', fontsize=11)
plt.ylabel('Activity (a.u.)', fontsize=11)
plt.xlim(xmin, xmax)
plt.ylim(0, ymax)
plt.xticks(range(xmin, xmax+1, 100))
plt.yticks(
    [round(y, 1) for y in range(0, int(ymax)+1, int(ymax/6))],
    [f"{y/1e11:.1f}" for y in range(0, int(ymax)+1, int(ymax/6))]
)
plt.tight_layout()
plt.savefig(f'{path}/raman_plot_proton.png', transparent=True)
plt.show()
```
